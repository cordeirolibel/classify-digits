# Weights

```python
from network import Network
from common import tic, separate
import mnist_loader as ml
from pca import Pca
import pandas as pd
import numpy as np 
from bokeh.io import output_notebook
from bokeh.plotting import figure, show, output_file
```

## Data:

```python
names = ['train_rot_pca50','test_pca50','validation_pca50']
df_train, df_validation, df_test = ml.load(names)
df_train.head()
```

## Net:

```python
net = Network([50,50, 10])
net.cost_function = 'cross-entropy'
net.regularization_parameter = 5.0
net2 = Network([50,50, 10],large_weight=True)
net2.cost_function = 'cross-entropy'
net2.regularization_parameter = 5.0
```

## Train:

```python
%%time
net.SGD(df_train, 500, 10, 0.1, df_test=df_test,save_ws=True)
```

```python
%%time
net2.SGD(df_train, 30, 10, 0.1, df_test=df_test,save_ws=True)
```

```python
print(len(net.df_weights))
net.df_weights.head()
```

## Plot:

```python
output_notebook()
plot = figure(title="Small Weights")
# first weights
w_array = net.df_weights.iloc[0].values
hist, edges = np.histogram(w_array, bins=40, density=True)
bincenters = 0.5*(edges[1:]+edges[:-1])
plot.line(bincenters,hist, line_width=2,legend='first',color="blue")
# last weights
w_array = net.df_weights.iloc[-1].values
hist, edges = np.histogram(w_array, bins=40, density=True)
bincenters = 0.5*(edges[1:]+edges[:-1])
plot.line(bincenters,hist, line_width=2,legend='last',color="red")

show(plot)
```

```python
output_notebook()
plot = figure(title="Large Weights")
# first weights
w_array = net2.df_weights.iloc[0].values
hist, edges = np.histogram(w_array, bins=40, density=True)
bincenters = 0.5*(edges[1:]+edges[:-1])
plot.line(bincenters,hist, line_width=2,legend='first',color="blue")
# last weights
w_array = net2.df_weights.iloc[-1].values
hist, edges = np.histogram(w_array, bins=40, density=True)
bincenters = 0.5*(edges[1:]+edges[:-1])
plot.line(bincenters,hist, line_width=2,legend='last',color="red")

show(plot)
```
